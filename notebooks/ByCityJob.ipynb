```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../datasets/FinalDF.csv')
df
```

### Population to numeric

```python
df['City_Population'] = df['City_Population'].str.replace(',', '').astype(float)
df['NeighbourCity_Population'] = df['NeighbourCity_Population'].str.replace(',', '').astype(float)
```

```python
df['TuitionFees'] = df['TuitionFees'].replace(['Null', 'null'], None)

df['TuitionFees'] = df['TuitionFees'].str.replace(',', '').astype(float)
```

```python
df.info()
```

###Feature Engineering

```python
df['Job_Density_Neighbor'] = df['Jobs in neighbour cities'] / (df['NeighbourCity_Population'] + 1e-6)

df['Connectivity_Score'] = df['Frequency'] / (df['Time'] * df['Distance (km)'] + 1e-6)

city_type_weights = {'Metropolitan City': 3, 'City': 2, 'Small Town': 1}
df['City_Importance_Score'] = df['NeighbourCity_Type'].map(city_type_weights)
```

```python
df
```

###Normalization

```python
from sklearn.preprocessing import MinMaxScaler

columns_to_scale = [
    'TuitionFees', 'Jobs in City', 'Jobs in neighbour cities', 
    'City_Population', 'NeighbourCity_Population', 'Time', 
    'Frequency', 'Distance (km)', 
    'Job_Density_Neighbor',
    'Connectivity_Score',
    'City_Importance_Score'

]

scaler = MinMaxScaler()
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
```

```python
df
```

###Features weighting

```python
weights = {
    'Jobs in City': 0.0,
    'Job_Density_Neighbor':0.1,
    'Connectivity_Score':0.7,
    'City_Importance_Score':0.2,    
}

df['Recommendation_Score'] = (
    weights['Jobs in City'] * df['Jobs in City'] +
    weights['Job_Density_Neighbor'] * df['Job_Density_Neighbor'] +
    weights['Connectivity_Score'] * df['Connectivity_Score'] +
    weights['City_Importance_Score'] * df['City_Importance_Score'] 

)
```

###University Recommendation

```python
recommended_universities = df.sort_values(by='Recommendation_Score', ascending=False)

recommended_universities[['CourseNameShort', 'Academy', 'City', 'Recommendation_Score']].head(20)
```

```python

weights = {
    'Jobs in City': 1,
    'Job_Density_Neighbor':2,
    'Connectivity_Score':6,
    'City_Importance_Score':1,    
}

def recommend_by_user_preference(weights, df):
    
    weights['Jobs in City'] = weights.get('Jobs in City') / 10
    weights['Job_Density_Neighbor'] = weights.get('Job_Density_Neighbor') / 10
    weights['Connectivity_Score'] = weights.get('Connectivity_Score') / 10
    weights['City_Importance_Score'] = weights.get('City_Importance_Score') / 10
    
    df['Recommendation_Score'] = (
    weights['Jobs in City'] * df['Jobs in City'] +
    weights['Job_Density_Neighbor'] * df['Job_Density_Neighbor'] +
    weights['Connectivity_Score'] * df['Connectivity_Score'] +
    weights['City_Importance_Score'] * df['City_Importance_Score']  
    )
    
    recommended_universities = df.sort_values(by='Recommendation_Score', ascending=False)

    recomendations = recommended_universities[['Academy', 'CourseNameShort', 'City', 'Recommendation_Score']].head(10)
    
    return recomendations
    
```

```python
result = recommend_by_user_preference(weights, df)
result.rename(columns={'CourseNameShort': 'Course Name', 'Academy': 'University', 'City': 'City', 'Recommendation_Score': 'Recommendation Score'}, inplace=True)
result.index = np.arange(1, len(result) + 1)
result['City'] = result['City'].str.title()
result['Recommendation Score'] = result['Recommendation Score'].apply(lambda x: round(x * 100, 2))
```
